```python
import urllib
import os
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import datetime
plt.rcParams['figure.figsize'] = 16, 20
```

```python
def create_time_grid(start_date='1/1/2018', end_date='2/28/2018'):
    df = pd.DataFrame(columns=['date', 'unixtime'])
    df.loc[:, 'date'] = pd.date_range(start=start_date, end=end_date)
    df.loc[:, 'unixtime'] = pd.to_datetime(df.loc[:, 'date']).astype(int)
    df.loc[:, 'unixtime'] = df.unixtime/1e9
    return df
```

```python
def get_weather_data(start_date='1/1/2018',
                     end_date='2/28/2018',
                     lat=13.0826802,
                     lon=80.2707184,
                     secret_key=''):
    this_time_grid = create_time_grid(start_date, end_date)
    df_list=[]
    for index, value in this_time_grid.iterrows():
        time = int(value.unixtime)
        print time
        this_url = 'https://api.darksky.net/forecast/' + secret_key + '/' + str(lat) + "," + str(lon) + "," + str(time) + '?'
        jsonString = urllib.urlopen(this_url).read()
        js = json.loads(jsonString)
        df = pd.DataFrame()
        for i in np.arange(24):
            this_dict = dict(js['hourly']['data'][i])
            dr = pd.DataFrame(this_dict, index=[i])
            df_list.append(dr)
    df_final = pd.concat(df_list, sort=False)
    return df_final
```
